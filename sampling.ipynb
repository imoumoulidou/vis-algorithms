We define the sample sizes for the experiment. Following the paper of 
'Evaluation of Sampling Methods for Scatterplots', we start with a small number of points and increase the sample size following a geometric series. As the sample size increase, we expect to observe less significant differences. 
In fact, prior works shows if we sample more than 50% of the dataset, then all sampling techniques start to match. 

We currently support the following sampling techniques:
    
    1. Random Sampling: ['random']
    2. Max-Min Sampling: ['max_min']
    3. Visualization-aware Sampling: ['vas']
    4. Density-biased Sampling: ['density_biased']
    5. Blue-noise Sampling: ['blue_noise']
    6. Perception-aware Sampling: ['perception_aware']
    
We show how to collect samples along with their saliency information for each sample size for these techniques. 

```python
sample_sizes = [250]

for i in range(1, 11, 2):
    sample_sizes.append(int(np.round(250*1.5**i)))
    
techniques = ['random', 'max_min', 'density_biased', 'perception_aware', 'vas', 'blue_noise']
seeds = np.array([0])
```

```python
print("Sample sizes", sample_sizes)
```

Code for generating the samples -- this is fast and stores the samples as npz files. 
This is what is needed to generate samples 

```python
for technique in techniques:
    produce_samples(dataset, technique, seeds, sample_sizes) # test for all sizes
    #produce_samples(dataset, technique, seeds, [sample_sizes[0]]) # test one size only
```

Code to generate saliency information for evaluation purposes. Run to generate the necessary information to quantitatively compare sampling methods

```python
"""
Warning: Generating the data we need to *evaluate* the sampling methods takes time & space. 
For *each* method and *each* sample size, we compute 16 stimuli that we feed into the DVS (saliency) model. 
Then for each stimulus, we get a saliency map that we first store into npz, and then plot as an image. 

All the evaluation metrics are computed on images, but currently our code stores the saliency maps as matrices 
as well from DVS. Future implementation will automatically clear any intermediate files that are not used. 
"""

for technique in techniques:
    produce_saliency_data(dataset, technique, seeds, sample_sizes)
    #produce_saliency_data(dataset, technique, seeds, [sample_sizes[0]])
```
