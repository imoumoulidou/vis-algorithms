We specify the configuration for the stimuli we generate to compute an aggregate saliency map. Spefically:

1. Fix the canvas size to 14 x 12 inch. 
2. Vary ppacity: [0.1, 0.4, 0.7, 1.0]
3. Vart point size: [2, 4, 8, 16]

The framework expects json files describing the experimetal setup. 

```python
# Code to set the experimental settings 
setting = {}

# Figure size 
setting['x_canvas_size'] = 14 
setting['y_canvas_size'] = 12 

# Resolution
setting['dpi_value'] = 100
setting['marker'] = 4
setting['alpha'] = 1 

# Limits while plotting the data onto a scatterplot. Guarantees that all plots are generated similarly
setting['x_lim'] = [-0.05, 1.05]
setting['y_lim'] = [-0.05, 1.05]

# Color for points in scatterplot 
setting['color'] = '#000C66' 
setting['background_color'] = 'white'

setting['resize_x'] = 800 
setting['resize_y'] = 640 

# Transparency values you would like to experiment with while plotting visual stimuli
transparency_values = np.arange(0.1, 1.3, 0.3) 

for i in range(transparency_values.shape[0]):
    transparency_values[i] = round(transparency_values[i], 3)

# Marker values you would like to experiment with while plotting visual stimuli
marker_values = np.array([2**m for m in range(1, 5)])

setting['transparency_values'] = transparency_values.tolist()
setting['marker_values'] = marker_values.tolist()

setting['marker_level_start'] = 0
setting['marker_level_end'] = marker_values.shape[0] 

setting['transparency_level_start'] = 0
setting['transparency_level_end'] = transparency_values.shape[0]

setting['epsilon'] = 1e-2

# Generating setting for perception-aware weighting scheme
with open(f"{dataset}/weighting_phase_setting.json", "w") as outfile:
    json.dump(setting, outfile)

# Generating setting for evaluation scheme 
with open(f"{dataset}/sampling_phase_setting.json", "w") as outfile:
    json.dump(setting, outfile)
```
